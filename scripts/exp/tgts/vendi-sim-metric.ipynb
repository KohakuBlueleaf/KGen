```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
best_sim = np.load("../../output/best-sims.npy", allow_pickle=True).tolist()
worst_sim = np.load("../../output/worst-sims.npy", allow_pickle=True).tolist()
model_order = [
    ("oai", "GPT4o-mini"),
    ("original", "Original"),
    ("promptdb", "Prompt-DB"),
    ("promptist", "Promptist"),
    ("tipo", "TIPO"),
]
print(best_sim)
```

```python
# Plot each matrix
for key, legend_name in model_order:
    plt.figure(figsize=(7, 6))
    sns.heatmap(
        best_sim[key] * 2 - 1,
        cmap="YlOrRd",
        vmin=0,
        xticklabels=False,
        yticklabels=False,
    )
    plt.title(legend_name)
    plt.tight_layout()
    plt.savefig(
        f"../../output/sim-mat/sim_matrix_{key}_best.png", dpi=300, bbox_inches="tight"
    )
    plt.close()
for key, legend_name in model_order:
    plt.figure(figsize=(7, 6))
    sns.heatmap(
        worst_sim[key] * 2 - 1,
        cmap="YlOrRd",
        vmin=0,
        xticklabels=False,
        yticklabels=False,
    )
    plt.title(legend_name)
    plt.tight_layout()
    plt.savefig(
        f"../../output/sim-mat/sim_matrix_{key}_worst.png", dpi=300, bbox_inches="tight"
    )
    plt.close()
```
