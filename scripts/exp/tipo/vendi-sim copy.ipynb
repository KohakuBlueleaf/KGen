```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
model_order = [
    ("original-best", "Original"),
    ("original-worst", "Original"),
    ("oai-best", "GPT4o-mini"),
    ("oai-worst", "GPT4o-mini"),
    ("promptdb-best", "MagicPrompt"),
    ("promptdb-worst", "MagicPrompt"),
    ("promptist-best", "Promptist"),
    ("promptist-worst", "Promptist"),
    ("tipo-best", "TIPO"),
    ("tipo-worst", "TIPO"),
]
sim = {
    key: np.load(f"../../../output/{key}-sim.npy", allow_pickle=True)
    for key, _ in model_order
}
print(sim.keys())
```

```python
# Plot each matrix
for key, legend_name in model_order:
    plt.figure(figsize=(7, 6))
    choosed = np.random.choice(range(len(sim[key])), min(256, len(sim[key])), False)
    sim = sim[key][choosed][:, choosed]
    print(np.min(sim))
    sns.heatmap(
        sim,
        cmap="YlOrRd",
        vmin=0.5,
        xticklabels=False,
        yticklabels=False,
    )
    plt.title(legend_name)
    plt.tight_layout()
    plt.savefig(
        f"../../../output/sim-mat/sim_matrix_{key}-test.png",
        dpi=300,
        bbox_inches="tight",
    )
    plt.close()
    break
```
